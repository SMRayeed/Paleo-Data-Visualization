```python
import os
import pandas as pd
from io import StringIO
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
files_path = '/content/drive/MyDrive/AI4C-Project/'
```

```python
files = os.listdir(files_path)
```

```python
for items in files :
  print(items)
  if "Oak" in items :
    files_path_oaks = files_path + items
  if "Maples" in items :
    files_path_maples = files_path + items
  if "Firs" in items :
    files_path_firs = files_path + items
```

```python
csv_files_path = "/Files(.csv)/"
```

```python
oaks_text_path = files_path_oaks + csv_files_path
maples_text_path = files_path_maples + csv_files_path
firs_text_path = files_path_firs + csv_files_path
```

```python
oaks_text = os.listdir(oaks_text_path)
maples_text = os.listdir(maples_text_path)
firs_text = os.listdir(firs_text_path)
```

#### Oaks

```python
drive_path_csv = files_path_oaks + '/Same-size(.csv)/'
os.makedirs(drive_path_csv) if not os.path.exists(drive_path_csv) else None
```

```python
years_range = range(1, 2024)

dfs = []


for item in oaks_text :

  df = pd.read_csv(oaks_text_path + item)

  for year in years_range:
    column_name = str(year)
    if column_name not in df.columns : df[column_name] = float('nan')

  df = df.reindex(sorted(df.columns, key=lambda x: int(x) if x.isdigit() else float('inf')), axis=1)
  non_year_columns = [col for col in df.columns if not col.isdigit()]
  column_order = ['index'] + sorted(list(map(str, range(1, 2024))), key=lambda x: int(x)) + non_year_columns
  df = df[column_order]
  column_index_to_drop = df.shape[1] - 8
  df = df.iloc[:, :column_index_to_drop].join(df.iloc[:, column_index_to_drop + 1:])

  assert len(df.columns.to_list()) == 2031 , "BUG"
  if len(df.columns.to_list()) != 2031 : print(item)

  output_csv_file_path = drive_path_csv + item
  df.to_csv(output_csv_file_path, index=False)
  dfs.append(df)
```

```python
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
merged_df
```

```python
merged_df.to_csv('/content/drive/MyDrive/AI4C-Project/Oak/merged_oaks.csv', index=False)
```
