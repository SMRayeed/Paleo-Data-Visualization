#### Species found in Sites : with Counts

```python
species_sites_counts = df.groupby(['species', 'site']).size().reset_index(name='count')
```

```python
print("{:<20}{:<80}{}".format("Species", "Site", "Count"))
print("="*105)
for _, row in species_sites_counts.iterrows():
    species = row['species']
    site = row['site']
    count = row['count']
    print("{:<20}{:<80}{}".format(species, site, count))
```

#### Sites with Multiple Species

```python
multiple_species_sites = species_sites_counts[species_sites_counts['site'].duplicated(keep=False)]

print("{:<80}{:<20}{}".format("Site", "Species", "Count"))
print("="*105)
for site in multiple_species_sites['site'].unique():
    site_data = multiple_species_sites[multiple_species_sites['site'] == site]
    for _, row in site_data.iterrows():
        site = row['site']
        species = row['species']
        count = row['count']
        print("{:<80}{:<20}{}".format(site, species, count))
    print('-'*105)
```

```python
import plotly.express as px

# Assuming multiple_species_sites is your DataFrame

species_colors = {'PIAL': 'rgb(31, 119, 180)',
                  'PIAZ': 'rgb(255, 127, 14)',
                  'PICO': 'rgb(44, 160, 44)',
                  'PIEC': 'rgb(214, 39, 40)',
                  'PIED': 'rgb(148, 103, 189)',
                  'PIJE': 'rgb(140, 86, 75)',
                  'PILA': 'rgb(227, 119, 194)',
                  'PIPA': 'rgb(127, 127, 127)',
                  'PIPO': 'rgb(249, 189, 34)',
                  'PIRE': 'rgb(23, 190, 207)',
                  'PISF': 'rgb(140, 205, 210)',
                  'PIST': 'rgb(114, 27, 35)',
                  'PITA': 'rgb(255, 152, 150)'}

fig = px.bar(multiple_species_sites, x='site', y='count', color='species',
             labels={'count': 'Count', 'site': 'Site'},
             category_orders={'site': multiple_species_sites['site'].unique()},
             width=800, height=500,
             color_discrete_map=species_colors)

fig.update_layout(xaxis_title='Site', yaxis_title='Count', legend_title='Species',
                  xaxis=dict(tickangle=270, tickmode='array', tickvals=list(range(len(multiple_species_sites['site'].unique()))),
                             ticktext=multiple_species_sites['site'].unique()),
                  bargap=0.5, bargroupgap=0.25)  # Adjust these values as needed

fig.show()
```
